```python
def to_float_euros(x):
    if pd.isna(x): 
        return np.nan
    s = str(x)
    s = s.replace("€","").replace(",",".")
    s = re.sub(r"[^\d\.\-]", "", s)
    try:
        return float(s)
    except:
        return np.nan

if 'depense_mensuelle_culturelle' in df.columns:
    df['depense_mensuelle_culturelle_num'] = df['depense_mensuelle_culturelle'].apply(to_float_euros)

# Coerce some numeric-like columns
for maybe_num in ['nb_enfants','taille_foyer']:
    if maybe_num in df.columns:
        df[maybe_num] = pd.to_numeric(df[maybe_num], errors='coerce')

# Convert age categories to numeric midpoints if age is categorical (e.g., "18-24", "25-34", "65+")
def age_to_midpoint(a):
    if pd.isna(a): 
        return np.nan
    s = str(a)
    if re.match(r"^\d+\s*-\s*\d+$", s):
        lo, hi = re.split(r"\s*-\s*", s)
        return (float(lo)+float(hi))/2.0
    mplus = re.match(r"^(\d+)\s*\+$", s)
    if mplus:
        return float(mplus.group(1)) + 5  # arbitrary midpoint for open-ended
    monly = re.match(r"^\d+$", s)
    if monly:
        return float(monly.group(0))
    return np.nan

if 'age' in df.columns:
    df['age_mid'] = df['age'].apply(age_to_midpoint)

# Helper aggregations
def topn_mean_by(group_col, value_col, n=15):
    tmp = df[[group_col, value_col]].dropna()
    g = tmp.groupby(group_col)[value_col].mean().sort_values(ascending=False).head(n)
    return g

def counts(col):
    return df[col].value_counts(dropna=False)

def percent_table(group_col, target_col):
    ctab = pd.crosstab(df[group_col], df[target_col], normalize='index') * 100
    return ctab
```

```python
# ---------- WHAT-IF (simple scenario function) ----------
# A simple function to simulate an uplift on spending and visualize static histogram
def simulate_spending_uplift(uplift_pct=10.0):
    if 'depense_mensuelle_culturelle_num' not in df.columns:
        print("La colonne de dépense n'est pas disponible.")
        return
    base = df['depense_mensuelle_culturelle_num'].dropna()
    uplifted = base * (1 + uplift_pct/100.0)
    plt.figure(figsize=(9,4))
    plt.hist(base, bins=40, alpha=0.6, label="Base")
    plt.hist(uplifted, bins=40, alpha=0.6, label=f"+{uplift_pct:.0f}%")
    plt.title(f"Simulation What-if : Histogrammes de dépense (base vs +{uplift_pct:.0f}%)")
    plt.xlabel("Dépense mensuelle (euros)")
    plt.ylabel("Fréquence")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
